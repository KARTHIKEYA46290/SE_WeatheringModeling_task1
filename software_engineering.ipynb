```python
import matplotlib.pyplot as plt
import numpy as np

def quadratic_model(time):
  # Hard coded coeffficients for the quadratic equation
  a=0.1
  b=-1
  c=30

  # Quadratic Equation representing temperature change
  temperature = a*time**2 + b*time + c

  return temperature

def main():
  #Time values from 0 to 10 (for example)
  time_values = np.linspace(0, 10, 50)

  #Calculate temperatures using hard-coded coefficients
  temperature_hardcoded = quadratic_model(time_values)

  #Plot the results
  plt.plot(time_values, temperature_hardcoded, label='Hard-coded Coefficients')
  plt.xlabel('Time')
  plt.ylabel('Temperature')
  plt.legend()
  plt.title('Weather Modeling with Quadratic Equation(Hard Coded Coefficient)')
  plt.show()

if __name__ == '__main__':
  main()
```

```python
import matplotlib.pyplot as plt
import numpy as np

def quadratic_model(time,a,b,c):
  # Hard coded coeffficients for the quadratic equation

  # Quadratic Equation representing temperature change
  temperature = a*time**2 + b*time + c

  return temperature

def main():
  #Time values from 0 to 10 (for example)
  time_values = np.linspace(0, 10, 50)
  with open ('/content/coefficients.txt/coefficients1.txt', 'r') as file:
    coefficients = []
    for line in file:
      a, b, c = map(float, line.split())
      coefficients.append([a, b, c])
  #Calculate temperatures using hard-coded coefficients
  for i , coefficient in enumerate(coefficients):
    a, b, c = coefficient
    temperature_hardcoded = quadratic_model(time_values,a,b,c)
    plt.plot(time_values, temperature_hardcoded, label=coefficients[i])

  #Plot the results
  plt.xlabel('Time')
  plt.ylabel('Temperature')
  plt.legend()
  plt.title('Weather Modeling with Quadratic Equation (Multiple set of Coefficients from file)')
  plt.show()

if __name__ == '__main__':
  main()
```
